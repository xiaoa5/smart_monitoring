```python
import torch
from gsplat import rasterization
import numpy as np
from PIL import Image
from plyfile import PlyData

# Load ply (your code)
plydata = PlyData.read("outputs/poster/splatfacto/2025-11-08_133853/point_cloud/splat.ply")
vertex_data = plydata['vertex'].data

# Extract positions, scales, rotations, opacity
means = torch.tensor(np.stack([vertex_data['x'], vertex_data['y'], vertex_data['z']], axis=1), dtype=torch.float32)
scales = torch.tensor(np.exp(np.stack([vertex_data['scale_0'], vertex_data['scale_1'], vertex_data['scale_2']], axis=1)), dtype=torch.float32)
quats = torch.tensor(np.stack([vertex_data['rot_0'], vertex_data['rot_1'], vertex_data['rot_2'], vertex_data['rot_3']], axis=1), dtype=torch.float32)
quats = quats / torch.norm(quats, dim=1, keepdim=True)  # Normalize
opacities = torch.sigmoid(torch.tensor(vertex_data['opacity'], dtype=torch.float32))  # (N,)

# Extract SH: DC + rest (for degree 3, 16 coeffs total per color)
f_dc = torch.tensor(np.stack([vertex_data['f_dc_0'], vertex_data['f_dc_1'], vertex_data['f_dc_2']], axis=1), dtype=torch.float32)
sh_degree = 0  # Default to DC only
shs = f_dc.unsqueeze(1)  # (N, 1, 3)

# Check for and extract higher SH if present (f_rest_0 to f_rest_44 for degree 3, packed as 15*3=45)
if 'f_rest_0' in vertex_data.dtype.names:
    num_rest = sum(1 for name in vertex_data.dtype.names if name.startswith('f_rest_'))
    if num_rest == 45:  # Expected for degree 3
        f_rest = np.stack([vertex_data[f'f_rest_{i}'] for i in range(45)], axis=1).astype(np.float32)
        sh_rest = torch.tensor(f_rest).reshape(-1, 15, 3)  # (N, 15, 3)
        shs = torch.cat([shs, sh_rest], dim=1)  # (N, 16, 3)
        sh_degree = 3

# Optional filter
mask = opacities > 0.1
means, scales, quats, opacities, shs = [t[mask] for t in [means, scales, quats, opacities, shs]]

# To device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
means, scales, quats, opacities, shs = [t.to(device) for t in [means, scales, quats, opacities, shs]]

# Debug shapes
print(f"means: {means.shape}, quats: {quats.shape}, scales: {scales.shape}, opacities: {opacities.shape}, shs: {shs.shape}")

# New camera setup: position at [2, 2, 2], looking at [0, 0, 0], up [0, 1, 0]
cam_pos = torch.tensor([-3.0, 5.0, 3.0], device=device)
look_at = torch.tensor([200.0, 10.0, 100.0], device=device)
up = torch.tensor([1.0, 1.0, 1.0], device=device)

# Compute view matrix (cam-to-world)
forward = look_at - cam_pos
forward = forward / torch.norm(forward)
right = torch.cross(up, forward)
right = right / torch.norm(right)
up = torch.cross(forward, right)
rot = torch.stack([right, up, forward], dim=0)
viewmat = torch.eye(4, device=device)
viewmat[:3, :3] = rot
viewmat[:3, 3] = cam_pos
viewmat = viewmat.unsqueeze(0)  # (1, 4, 4)

# Intrinsics (perspective, FOV=60)
width, height = 800, 600
fov_deg = 60.0
fov_rad = torch.deg2rad(torch.tensor(fov_deg / 2.0))
fx = fy = width / (2.0 * torch.tan(fov_rad))  # Assuming square pixels; adjust if needed
K = torch.tensor([[fx, 0, width / 2], [0, fy, height / 2], [0, 0, 1]], device=device).unsqueeze(0)  # (1, 3, 3)

# Rasterize
rendered, alpha, _ = rasterization(
    means=means,  # (N, 3)
    quats=quats,  # (N, 4)
    scales=scales,  # (N, 3)
    opacities=opacities,  # (N,)
    colors=shs,  # (N, (sh_degree+1)**2, 3)
    viewmats=viewmat,
    Ks=K,
    width=width,
    height=height,
    sh_degree=sh_degree  # Auto-detects if 0
)

# Extract RGB (C=1, so [0])
rgb = rendered[0].cpu().numpy()  # (H, W, 3)
rgb = np.clip(rgb, 0, 1)  # Ensure [0,1]

# Save
Image.fromarray((rgb * 255).astype(np.uint8)).save('rendered_3dgs_new_view.png')
print("Saved as rendered_3dgs_new_view.png")
```
